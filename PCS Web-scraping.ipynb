```python
import requests
from bs4 import BeautifulSoup
```

```python
import pandas as pd
import numpy as np
```

```python
import csv
```

```python
url = "https://www.lolpcs.com/stats.php?Lan=1" 
html = requests.get(url)

url_source_code = html.text # get source code (stored in html.text)
soup = BeautifulSoup(url_source_code,"html.parser") # parse source code
```

#### save in list

```python
champion_stat_l = []

for hero in soup.find_all('div',class_="ranking-box"):
    
    try:
        hero_name = hero.find("div", class_="ranking-name hero").text.strip()       
        hero_rate = hero.find("div", class_="ranking-data").text.strip()
            
        champion_stat_l.append([hero_name,hero_rate])

    except:
        pass
```

```python
champion_stat_l
```

##### Ban Rate - save in CSV

```python
headings = ["Hero","Ban Rate"]
```

```python
with open('PCS Ban rate.csv', 'w', newline="") as output:
    writer = csv.writer(output)
    writer.writerow(headings)
    writer.writerows(champion_stat_l[:10])
```

### Scrape Players stats

```python
player_stat = soup.find_all('tr')

# check is it what I want
head = player_stat[0] 
body_rows = player_stat[1:]
```

```python
head
```

```python
body_rows
```

```python
# Scrape table column name

headings = []

for item in head.find_all("th"): # loop through all th elements
    headings.append(item.text)
    
print(headings)
```

```python
# Scrape all table data

rows = []

for row in player_stat:
    player_data = row.find_all('td') # loop through all th elements
    cleaned_player_data = [data.text.strip() for data in player_data]
   
    # Remove data arrays that are empty
    if len(cleaned_player_data) == 0:
        continue
        
    rows.append(cleaned_player_data)
```

```python
rows
```

#### save in CSV

```python
with open('PCS Players Stats.csv', 'w', newline="") as output:
    writer = csv.writer(output)
    writer.writerow(headings)
    writer.writerows(rows)
```
